```python
from keras.models import load_model
from keras.preprocessing import image

model = load_model("/content/gdrive/MyDrive/Monument Classifier/model.h5")
```

```python
sample = image.load_img("/content/gdrive/MyDrive/Monument Classifier/datasets/train/Aga Khan Palace/Aga Khan Palace24.jpg")
x = image.img_to_array(sample)
x = np.expand_dims(x, axis=0)

images = np.vstack([x])
pred = model.predict(images)
```

```python



classes = {
    15 : 'india gate',
    23 : 'india gate',
    37 : 'Agha Khan Palace',
    44 : 'hindu temple',
    48 : 'hindu temple',

    3 : 'bhudha temple',
    1 : 'Badrinath temple',
    31 : 'charminar',
    66 : "statue of unity"

}


def predict_image(img):
  # test = cv2.imread(img)
  # test = cv2.resize(test, (64,64))
  # test = np.reshape(test, [1,64,64,3])
  img = img.reshape((-1, 64, 64, 3))
  predict_x = model.predict(img)

  class_ = int(np.argmax(predict_x, axis=1))
  # prediction=class_
  print("CLasas ============= ", class_)

  # print(prediction)
  # return {classes[i]: float(prediction[i]) for i in range(5)}
  return classes.get(class_)

# image = gr.inputs.Image(shape=(64,64))

# label = gr.outputs.Label(num_top_classes=5)

gr.Interface(fn=predict_image, inputs="image", outputs=gr.Label(num_top_classes=49)).launch(debug='True', share=True)
# gr.Interface(fn=predict_image, inputs="image", outputs="label").launch(debug='True', share=True)
```
